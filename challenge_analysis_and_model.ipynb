# Building Model

```python
#choosing a route and extracting data for this route
route_name = "Sydney-Singapore"
route_df = df[df["Route"] == route_name]
monthly_series = route_df.groupby(["Date", "Route"])["Passengers_Total"].sum().reset_index()
monthly_series
```

```python
# creating and analyzing route's plot
plt.plot(monthly_series["Date"], monthly_series["Passengers_Total"])
plt.xticks(rotation=90)
```

```python
# splitting data - train/test dataframes
train = monthly_series.loc[monthly_series["Date"] < '1988-06-01']
test = monthly_series.loc[monthly_series["Date"] >= '1988-06-01']
```

```python
# creates a copy of df with new data-based features added
def create_features(df):
    df = df.copy()
    df["Month"] = df["Date"].dt.month
    df["Quarter"] = df["Date"].dt.quarter
    df["Year"] = df["Date"].dt.year
    
    return df
```

```python
train = create_features(train)

test = create_features(test)
```

```python
# creating features/target dataframes
feature_cols = ["Month", "Quarter", "Year"]

X_train = train[feature_cols]
y_train = train["Passengers_Total"]

X_test = test[feature_cols]
y_test = test["Passengers_Total"]
```

```python
import xgboost as xgb

# creating a model using XGBRegressor
model = xgb.XGBRegressor(n_estimators=1000, random_state=42)
model.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)],
    verbose=True)

y_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

```python
# visualising model performance

plt.plot(train["Date"], y_train, label="Train Actual", color='blue')
plt.plot(train["Date"], y_train_pred, label="Train Predicted", color='cyan', linestyle='--')
plt.plot(test["Date"], y_test, label="Test Actual", color='orange')
plt.plot(test["Date"], y_pred, label="Test Predicted", color='red', linestyle='--')
plt.xticks(rotation=90)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

# evaluating model performance
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"Test RMSE: {rmse:,.0f}")
print(f"Test R²: {r2:,.3f}")
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# creating a model using ExponentialSmoothing
model = ExponentialSmoothing(y_train, trend="add", seasonal="add", seasonal_periods=12)
fit = model.fit()
forecast = fit.forecast(13)
```

```python
# visualising model performance
plt.plot(train["Date"], train["Passengers_Total"], label="Train")
plt.plot(test["Date"], test["Passengers_Total"], label="Test")
plt.plot(test["Date"], forecast, label="Forecast")
plt.legend()
plt.xticks(rotation=90)
```

```python
# model evaluation
y_pred_2 = forecast[:len(y_test)]

rmse = mean_squared_error(y_test, y_pred_2, squared=False)
r2 = r2_score(y_test, y_pred_2)
print(f"Test RMSE: {rmse:.0f}")
print(f"Test R²: {r2:.3f}")
```
